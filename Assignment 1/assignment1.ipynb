## Data Visualization Using a Health Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Observing Data:

```python
df=pd.read_csv("Health_heart_experimental.csv")
df.head(10)
```

### Line Plot:

```python
sns.lineplot(x='BMI',y='HR',data=df)
```

##### In the above line graph, we see people with BMI from 25 to 30 having high heart rates. This indicates that their hearts work harder which stresses the cardiovascular system. The data indicate that the people from whom the data was collected have cardiovascular issues as their heart rate seems to be over 135 bpm. Furthermore, the highest heart rate recorded among all the patients is almost 140 bpm which indicates a potential health risk. 

### Histogram:

```python
plt.hist(df['BMI'])
```

##### In the above histogram, we can see that the histogram is left-skewed which indicates that the number of patients with BMI over 25 amount more in number than patients with BMI below 25. However, we can see that the number of patients with BMI of about 30 amount less in number which suggests that the trend is not strictly followed and the data seem to have been collected from different sources. Almost 10,000 patients seem to have a BMI closer to 35 which indicates serious health risk.

### Bar Plot:

```python
sns.barplot(x='sex',y='BMI',data=df)
```

##### When we look at the bar plot for patients of different sex and compare their BMI levels, it seems that the average BMI is same. This indicates that sex does not significantly impact one's BMI. However, there may be slight differences that the bar plot has failed to take into account since we are looking at the average BMI levels and not BMI levels across the board.

### Box Plot:

```python
sns.boxplot(df['SysBP'])
```

##### In the box plot above, we can see that the median value for Systolic Blood Pressure lies between 120 and 130. This means there are equal number of patients with hypertension (high blood pressure) and normal blood pressure. It seems that the patients have a systolic blood pressure ranging from about 115 to about 135. A systolic blood pressure of below 120 indicates normal blood pressure. Anything higher than 130, however, can be a health hazard. 

### Pie chart:

```python
df['indication'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

##### Above is the  pie chart of Healthy BMI range (0) & Unhealthy BMI range (1). BMI levels above 18.5 and below 25  are generally considered to fall within Healthy BMI range and anything above 25 is considered to be unhealthy. We can see that the given dataset consists of more patients with healthy BMI levels i.e. 52.7%. It seems that almost 47.3% of patients have a BMI value above 25 which indicates that their cardiovascular system is working harder to pump blood throughout their body.
